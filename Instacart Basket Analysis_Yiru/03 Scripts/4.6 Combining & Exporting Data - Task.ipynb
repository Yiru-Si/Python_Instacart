## Step 5: Determine a suitable way to combine the orders_products_combined dataframe with your products data set. 

```python
# Import products data
df_prods_checked = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col = False)
```

```python
df_prods_checked.head()
```

```python
df_orders_products_combined.head()
```

```python
# Drop 'unnamed' column in the products data set
df_prods_checked = df_prods_checked.drop(columns = ['Unnamed: 0'])
```

```python
# Re-check the products data set for the dropped column
df_prods_checked.head()
```

```python
# Drop '_merge' column in the orders_products_combined dataframe: the coming merging will create the same column name, which will cause error.
df_orders_products_combined = df_orders_products_combined.drop(columns = ['_merge'])
```

```python
# Re-check the orders_products_combined dataframe for the dropped column
df_orders_products_combined.head()
```

```python
# Merge the products data set with orders_products_combined dataframe
df_ords_prods_merged = df_orders_products_combined.merge(df_prods_checked, on = 'product_id', indicator = True)
```

```python
df_ords_prods_merged.head()
```
