### Step 2: Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column using a suitable method.

```python
# To identify how busy each day of the week is: the frequency of orders of each day
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create a new column
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value == 0 or value == 1:
        result.append("Busiest days")
    elif value == 4 or value == 3:
        result.append("Slowest days")
    else:
        result.append("Regular days")
```

```python
result
```

```python
# Create a new column 'busiest_day_2' within the dataframe and set it equal to result
df_ords_prods_merge['busiest_day_2'] = result
```

### Step 4: When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# To identify how busy each hour of the day is: the frequency of orders of each day
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create a new column
result = []

for value in df_ords_prods_merge["order_hour_of_day"]:
    if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
        result.append("Most orders")
    elif value == 3 or value == 4 or value == 2 or value == 5 or value == 1 or value == 0 or value == 6 or value == 23:
        result.append("Fewest orders")
    else:
        result.append("Average orders")
```

```python
# Create a new column 'busiest_period_of_day' within the dataframe and set it equal to result
df_ords_prods_merge['busiest_period_of_day'] = result
```

```python
# Check whether the new column is in the dataframe 
df_ords_prods_merge.head()
```
